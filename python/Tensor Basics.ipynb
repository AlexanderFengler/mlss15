# Basic tensor operations

```python
import numpy as np
import matplotlib.pylab as plt
from scipy.io import loadmat
%matplotlib inline 
```

```python
# Load amino
amino=loadmat('datasets/amino.mat');
X=amino['X']
sz=tuple([int(n) for n in amino['DimX'][0]])
```

```python
# Check the dimensions
X.shape
```

```python
# Slicing
X[0,:,:].shape
```

```python
plt.rcParams['figure.figsize']=(10,5)
for kk in range(5):
    plt.subplot(1,5,kk+1)
    plt.imshow(X[kk,:,:])
```

```python
# Mode-1 unfolding
X1=X.reshape((5, 201*61)); X1.shape
```

```python
# Mode-2 unfolding
X2=X.transpose((1,2,0)).reshape((201, 61*5)); X2.shape
```

```python
# Mode-3 unfolding
X3=X.transpose((2,0,1)).reshape((61, 5*201)); X3.shape
```

```python
# Plot the singular values
plt.rcParams['figure.figsize']=(10,5)
plt.subplot(1,3,1); plt.plot(np.linalg.svd(X1,compute_uv=False),'-x')
plt.subplot(1,3,2); plt.plot(np.linalg.svd(X2,compute_uv=False),'-x')
plt.subplot(1,3,3); plt.plot(np.linalg.svd(X3,compute_uv=False),'-x')
```
